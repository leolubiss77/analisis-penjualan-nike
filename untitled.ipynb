```python
import pandas as pd
import os

file_path = 'Nike_Sales_Uncleaned.csv'

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("Lima baris pertama data:")
    print(df.head())

    print("\nInformasi umum tentang data (tipe data, non-null count):")
    print(df.info())

    print("\nStatistik deskriptif untuk kolom numerik:")
    print(df.describe())

    print("\nJumlah baris dan kolom:")
    print(df.shape)
else:
    print(f"Error: File '{file_path}' tidak ditemukan. Pastikan sudah diekstrak atau diunggah.")
```

```python
# Pastikan 'Order_Date' diubah ke tipe datetime
# 'errors='coerce'' akan mengubah nilai yang tidak bisa dikonversi menjadi NaT (Not a Time)
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')

# Cek berapa banyak nilai NaT (yang gagal dikonversi) setelah proses ini
print(f"Jumlah nilai 'NaT' di Order_Date setelah konversi: {df['Order_Date'].isnull().sum()}")

# Karena Order_Date adalah kunci waktu untuk analisis, kita akan menghapus baris yang Order_Date-nya NaT
# Ini lebih baik daripada mengisi tanggal yang tidak diketahui karena bisa mengacaukan analisis tren.
df.dropna(subset=['Order_Date'], inplace=True)

print(f"Jumlah baris setelah menghapus NaT di Order_Date: {df.shape[0]}")

# Periksa lagi tipe data Order_Date
print("\nInformasi tipe data setelah membersihkan Order_Date:")
print(df.info())
```

```python
# Ubah nilai negatif di 'Units_Sold' menjadi 0
# Gunakan .loc untuk menghindari SettingWithCopyWarning
df.loc[df['Units_Sold'] < 0, 'Units_Sold'] = 0
print(f"Jumlah 'Units_Sold' negatif setelah dibersihkan: {df[df['Units_Sold'] < 0].shape[0]}")

# Ubah nilai negatif di 'Revenue' menjadi 0
df.loc[df['Revenue'] < 0, 'Revenue'] = 0
print(f"Jumlah 'Revenue' negatif setelah dibersihkan: {df[df['Revenue'] < 0].shape[0]}")

# Cek kembali statistik deskriptif untuk memastikan nilai min sudah 0 atau positif
print("\nStatistik deskriptif setelah penanganan nilai negatif di Units_Sold dan Revenue:")
print(df[['Units_Sold', 'Revenue', 'Profit']].describe())
```

```python
# Isi missing values untuk kolom numerik dengan median
# Median lebih robust terhadap outliers dibanding mean
median_units_sold = df['Units_Sold'].median()
median_mrp = df['MRP'].median()
median_discount = df['Discount_Applied'].median()

df['Units_Sold'].fillna(median_units_sold, inplace=True)
df['MRP'].fillna(median_mrp, inplace=True)
df['Discount_Applied'].fillna(median_discount, inplace=True)

# Isi missing values untuk kolom 'Size' dengan modus (nilai paling sering muncul)
mode_size = df['Size'].mode()[0]
df['Size'].fillna(mode_size, inplace=True)


print("\nJumlah nilai hilang per kolom setelah penanganan:")
print(df.isnull().sum())

print("\nInformasi umum data setelah semua penanganan:")
print(df.info())
```

```python
# Asumsi: Discount_Applied seharusnya tidak lebih dari 1 (100%)
# Jika ada nilai > 1, kita bisa membatasi menjadi 1.0
df.loc[df['Discount_Applied'] > 1.0, 'Discount_Applied'] = 1.0

print("\nStatistik deskriptif untuk Discount_Applied setelah penanganan anomali:")
print(df['Discount_Applied'].describe())
```

```python
# Ekstrak Tahun, Bulan, Hari, Hari dalam Seminggu, dan Jam dari Order_Date
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Month'] = df['Order_Date'].dt.month_name() # Nama bulan (misal: 'January')
df['Order_Day'] = df['Order_Date'].dt.day
df['Order_Day_of_Week'] = df['Order_Date'].dt.day_name() # Nama hari (misal: 'Monday')
df['Order_Hour'] = df['Order_Date'].dt.hour

# Tampilkan beberapa baris pertama dengan kolom baru
print("Lima baris pertama data setelah ekstrasi fitur waktu:")
print(df[['Order_Date', 'Order_Year', 'Order_Month', 'Order_Day', 'Order_Day_of_Week', 'Order_Hour']].head())
```

```python
# Hitung 'Total_Price' per item setelah diskon
# Harga_final = MRP * (1 - Discount_Applied)
df['Price_After_Discount'] = df['MRP'] * (1 - df['Discount_Applied'])

# Hitung 'Total_Sales' per baris (Total_Price * Units_Sold)
# Ini adalah pendapatan bersih dari penjualan item di baris tersebut
df['Total_Sales'] = df['Price_After_Discount'] * df['Units_Sold']

# Tampilkan beberapa baris pertama dengan kolom baru
print("\nLima baris pertama data setelah perhitungan harga dan total penjualan:")
print(df[['MRP', 'Discount_Applied', 'Price_After_Discount', 'Units_Sold', 'Total_Sales', 'Revenue']].head())

# Periksa statistik deskriptif untuk Total_Sales
print("\nStatistik deskriptif untuk Total_Sales:")
print(df['Total_Sales'].describe())

# Perhatikan bahwa ada perbedaan antara 'Revenue' asli dan 'Total_Sales' yang kita hitung.
# 'Revenue' asli mungkin sudah mencakup perhitungan kompleks lainnya atau merupakan nilai final yang dilaporkan.
# 'Total_Sales' ini adalah pendapatan yang kita hitung berdasarkan MRP, Diskon, dan Units_Sold.
# Kita akan menggunakan 'Total_Sales' untuk analisis kita, kecuali ada instruksi lain.
```

```python
# Gabungkan Gender_Category dan Product_Line untuk membuat segmen sederhana
df['Customer_Product_Segment'] = df['Gender_Category'] + ' - ' + df['Product_Line']

print("\nJumlah unik Customer_Product_Segment:")
print(df['Customer_Product_Segment'].nunique())
print(df['Customer_Product_Segment'].value_counts().head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Atur gaya plot
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Agregasi Total_Sales per bulan-tahun
# Urutkan berdasarkan Order_Year dan Order_Date (untuk memastikan urutan bulan yang benar)
sales_by_month_year = df.groupby(df['Order_Date'].dt.to_period('M'))['Total_Sales'].sum().reset_index()
sales_by_month_year['Order_Date'] = sales_by_month_year['Order_Date'].astype(str) # Ubah ke string untuk plotting

sns.lineplot(data=sales_by_month_year, x='Order_Date', y='Total_Sales', marker='o')
plt.title('Total Penjualan per Bulan-Tahun')
plt.xlabel('Bulan-Tahun')
plt.ylabel('Total Penjualan')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Kita juga bisa melihat penjualan per hari dalam seminggu
plt.figure(figsize=(10, 5))
sales_by_day_of_week = df.groupby('Order_Day_of_Week')['Total_Sales'].sum().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])
sns.barplot(x=sales_by_day_of_week.index, y=sales_by_day_of_week.values, palette='viridis')
plt.title('Total Penjualan per Hari dalam Seminggu')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Total Penjualan')
plt.tight_layout()
plt.show()
```
